# Bulid a RNN/ LSTM Model

# Import Packages

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# Create Data set
Data = [[[i+j] for i in range(5)] for j in range(100)]
target = [(i+5) for i in range(100)]
```

```python
# Nornmalize
Data = [[[(i+j)/100] for i in range(5)] for j in range(100)]
target = [(i+5)/100 for i in range(100)]
```

```python
# Check Datset
Data
```

```python
# Check target
target
```

```python
# Check dependent units
Data[:6]
```

```python
target[:6]
```

```python
# Convert Data to numpy arrays
data = np.array(Data, dtype=float)
```

```python
data.shape
```

```python
# Convert target into numpy arrays
target = np.array(target, dtype=float)
```

```python
target.shape
```

```python
# Split data train and test
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size = 0.3, random_state=4)
```

```python
# Check data
print(x_train.shape)
print(x_test.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
# RNN Model
model = Sequential()
```

```python
model.add(LSTM((1),batch_input_shape=(None,5,1),return_sequences= True))
model.add(LSTM((1),return_sequences= False))
```

```python
# Compile model
model.compile(loss= 'mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
# model summary
model.summary()
```

```python
# Fit the model
history = model.fit(x_train,y_train, epochs=500, validation_data=(x_test,y_test))
```

```python
# Model Prediction
result = model.predict(x_test)
result.shape
```

```python
y_test.shape
```

```python
# Plot the results
plt.scatter(range(30),result, c='r')
plt.scatter(range(30),y_test, c='g')
```
